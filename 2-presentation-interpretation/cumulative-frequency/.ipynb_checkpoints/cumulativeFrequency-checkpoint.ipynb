```python
import matplotlib.pyplot as plt
import numpy as np

# Grouped data
classes = [40, 50, 60, 70, 80]
frequencies = [2, 5, 3, 3]

# Compute cumulative frequencies
cumulative_frequencies = np.cumsum(frequencies)

# Plot
plt.plot(classes[1:], cumulative_frequencies, marker='o', linestyle='-', color='blue')
plt.title('Cumulative Frequency Graph')
plt.xlabel('Class Upper Bound')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [
    ["Height of Tree", "Class Boundaries", "Number of Trees ", "Cumulative"],
    ["0-5",   [0, 5.5],    26, 26],
    ["6-10",  [5.5, 10.5],    17, 43],
    ["11-15", [10.5, 15.5],   11, 54],
    ["16-20", [15.5, 20.5],   6, 60]
]

classes = [row[0] for row in data[1:]]
frequencies = [row[2] for row in data[1:]]
cumulative_frequencies = np.cumsum(frequencies)


print(classes)
print(frequencies)
print(cumulative_frequencies)

# Plot
plt.plot(classes, cumulative_frequencies, marker='o', linestyle='-', color='blue')
plt.title('Cumulative Frequency Graph')
plt.xlabel('Class Upper Bound')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```
