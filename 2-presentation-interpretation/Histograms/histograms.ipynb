```python
Data = [
    # [lower, upper, class width, Frequency, FrequencyDensity]
    [130, 150, 0, 3, 0],
    [150, 160, 0, 4, 0],
    [160, 165, 0, 5, 0],
]
```

```python
# Find Class Width

# Class Width = Upper - Lower

for Class in Data:
    Class[2] = Class[1] - Class[0]
print(Data)
```

```python
# Find FreqencyDensity

# Freqency Density = Frequency / Class Width

for Class in Data:
    Class[4] = Class[3] / Class[2]
print(Data)
```

```python
import matplotlib.pyplot as plt


freq_density = [ClassItem[4] for ClassItem in Data] 

# Calculate class widths
widths = [ClassItem[2] for ClassItem in Data]


# Use midpoints or left edges for bar positions
'''
 boundaries[:-1] means:
"All elements of the boundaries list except the last one."
'''
left_edges = [ClassItem[0] for ClassItem in Data]

# Plot histogram bars using bar()
plt.bar(left_edges, freq_density, width=widths, align='edge', edgecolor='black')

# Labels and title
plt.xlabel('Class Interval')
plt.ylabel('Frequency Density')
plt.title('Histogram from Grouped Data')

plt.show()
```
