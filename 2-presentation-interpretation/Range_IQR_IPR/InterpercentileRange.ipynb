```python
# Position of Xth percentile = x/100 * n
```

```python
data = [
    ["Height of Tree", "Class Boundaries", "Number of Trees ƒ", "Cumulative"],
    ["0-5",   [0, 5.5],    26, 26],
    ["6-10",  [5.5, 10.5],    17, 43],
    ["11-15", [10.5, 15.5],   11, 54],
    ["16-20", [15.5, 20.5],   6, 60]
]
```

```python
# 10th Percentile

Percentile = 10

Px = (Percentile / 100) * data[-1][3]

print(f'Position of {Percentile}th Percentile == {Px}')
```

```python
# Class of Percentile

PercentileClass = [] # Last index is used to find the one before that

print(f'Taking Rounded Q1 Index')
for Check in range(1,len(data)):
    if Px < data[Check][3]: # Check position with respects to cumulative value
        print(f'Percentile is in {data[Check][0]} Class')
        PercentileClass = data[Check]
        PercentileClass.append(Check) # Last index is used to find the one before that
        break
        
print(f'PercentileClass===={PercentileClass}')
```

```python
'''
LINEAR INTERPOLATION VIA RATIOS

b1===================
Lower Bound of PercentileClass ----- Percentile ----- Upper Bound of PercentileClass
a1=========

b2====================
Cumu Class -1 ----- Px ----- Cumu Percentile Class
a2=========

a1/b1 = a2/b2

the two ratios are the same because the items are assumed to be spread out evenly throughout the whole class

a1 = from Lower Bound of PercentileClass to Percentile --> Percentile - Lower Bound of PercentileClass
b1 = from Lower Bound of PercentileClass to Upper Bound of PercentileClass --> Upper Bound of PercentileClass - Lower Bound of PercentileClass
a2 = Cumu Class -1 Item to Px Item (position wise), Cumu Class -1 to Q1's position --> Px - cumu of Q1 -1 class
b2 = Cumu Class -1 to Cumu Percentile Class --> Cumu Percentile Class - Cumu Class -1

(Percentile - Lower Bound of PercentileClass) / (Upper Bound of PercentileClass - Lower Bound of PercentileClass) = (Px - Cumu Class -1 ) / (Cumu Percentile Class - Cumu Class -1)

Percentile - aOneWithoutPercentile / bOne = aTwo / bTwo

Percentile - aOneWithoutPercentile = ( aTwo / bTwo ) * bOne

Percentile = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutPercentile

'''



CumuPerMinus1 = 0

if PercentileClass[4] == 1:
    #This means the Q1 class was the first class
    CumuPerMinus1 = 0 # Start is 0
else:
    CumuPerMinus1 = data[PercentileClass[4] -1][3]

aOneWithoutPercentile = PercentileClass[1][0]
bOne = PercentileClass[1][1] - PercentileClass[1][0]
aTwo = Px -  CumuPerMinus1 # PercentileClass[4] has the index of the Percentile class, -1 is the one before that
bTwo = PercentileClass[3] -  CumuPerMinus1

# print(aTwo)

Percentile = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutPercentile

print(f'Percentile === {Percentile}')
```
