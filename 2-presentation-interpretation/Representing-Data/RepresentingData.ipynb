```python
# Grouped Frequency


# Inequality borders

WeightI = [
    ["W < 10",10],
    ["10 <= W < 100",21],
    [" 100 <= W",22]
]

# Number borders

Weight = [
    # [start, end, freq]
    [1,10,21],
    [11,20,13],
    [21,30,31]
]

'''
Notice for number, The lowest one that must be in the middle class is 10.5, as it'll be rounded up, 
and the highest is also the lowest of the next class, 20.5

think upper and lower bounds
'''
```

```python
for WeightClass in Weight:
    '''
    Class width = upper Class boundary - Lower Class Boundary
    '''
    print(f"Upper = {((WeightClass[1] + (WeightClass[1] + 1))/2)}")
    print(f"Lower = {((WeightClass[0] + (WeightClass[0] - 1))/2)}")
    # print(f"Lower = {(WeightClass[0])}")
    print(f"Class Width = {((WeightClass[1] + (WeightClass[1] + 1))/2) - ((WeightClass[0] + (WeightClass[0] - 1))/2)}")
```

```python
for WeightClass in Weight:
    '''
    Midpoint = (Lower Class Boundary + Upper Class Boundary) / 2
    '''
    print(f"Upper = {((WeightClass[1] + (WeightClass[1] + 1))/2)}")
    print(f"Lower = {((WeightClass[0] + (WeightClass[0] - 1))/2)}")
    # print(f"Lower = {(WeightClass[0])}")
    print(f"MidPoint = {( ((WeightClass[0] + (WeightClass[0] - 1))/2) + ((WeightClass[1] + (WeightClass[1] + 1))/2) ) / 2}")
```

```python
# Create new full array with all info

FullDataSet = []

for Record in Weight:
    toPush = Record
    toPush.append(( ((Record[0] + (Record[0] - 1))/2) + ((Record[1] + (Record[1] + 1))/2) ) / 2)
    FullDataSet.append(toPush)

print(FullDataSet)
```

```python
WeightFreq = []
WeightMidPoints = []

for Record in FullDataSet:
    WeightFreq.append(Record[2])
    WeightMidPoints.append(Record[3])
```

```python

# Plot
plt.plot(WeightMidPoints, WeightFreq, marker='o')  # 'marker' adds points
plt.xlabel('MidPoints of Weight Class')
plt.ylabel('Frequency')
plt.title('Dataset of FullDataSet')
plt.grid(True)
plt.show()
```
