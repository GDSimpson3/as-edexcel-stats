```python
data = [
    ["Height of Tree", "Class Boundaries", "Number of Trees ƒ", "Cumulative"],
    ["0-5",   [0, 5.5],    26, 26],
    ["6-10",  [5.5, 10.5],    17, 43],
    ["11-15", [10.5, 15.5],   11, 54],
    ["16-20", [15.5, 20.5],   6, 60]
]
```

```python
# Find Q1

import math

Q1Position = math.ceil(data[-1][3] / 4) # n

print(f'Q1Position==={Q1Position}')

# Q1 Class

Q1Class = [] # Last index is used to find the one before that

print(f'Taking Rounded Q1 Index')
for Check in range(1,len(data)):
    if Q1Position < data[Check][3]: # Check position with respects to cumulative value
        print(f'Q1 is in {data[Check][0]} Class')
        Q1Class = data[Check]
        Q1Class.append(Check) # Last index is used to find the one before that
        break
        
print(f'Q1Class===={Q1Class}')

'''
LINEAR INTERPOLATION VIA RATIOS

b1===================
-0.5 ----- Q1 ----- 5.5
a1=========

b2====================
0 ----- n/4 ----- 26
a2=========

a1/b1 = a2/b2

the two ratios are the same because the items are assumed to be spread out evenly throughout the whole class

a1 = from Q1 class's lower bound to Q1 --> Q1 - lower bound
b1 = from Q1 class's lower bound to Q1 class's Upper bound --> Upper bound - Lower Bound
a2 = 0th Item to Q1 Item (position wise), Cumulative of lower class to Q1's position --> Q1 Position - cumu of Q1 -1 class
b2 = Cumu of Q1 class -1 to Cumu of Q1 Class --> Cumu Q1 - (Cumu Q1 -1 class)

Q1 - Q1Lb / Q1Ub - Q1Lb = (Q1Position - CumuQ1-1 ) / (CumuQ1 - CumuQ1-1)

Q1 - aOneWithoutQ1 / bOne = aTwo / bTwo

Q1 - aOneWithoutQ1 = ( aTwo / bTwo ) * bOne

Q1 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ1

'''



CumuQ1Minus1 = 0

if Q1Class[4] == 1:
    #This means the Q1 class was the first class
    CumuQ1Minus1 = 0 # Start is 0
else:
    CumuQ1Minus1 = data[Q1Class[4] -1][3]

aOneWithoutQ1 = Q1Class[1][0]
bOne = Q1Class[1][1] - Q1Class[1][0]
aTwo = Q1Position -  CumuQ1Minus1 # Q1Class[4] has the index of the Q1 class, -1 is the one before that
bTwo = Q1Class[3] -  CumuQ1Minus1

# print(aTwo)

Q1 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ1

print(f'Q1 === {Q1}')
```

```python
# Find Q2

import math

Q2Position = math.ceil(data[-1][3] / 2) # n

print(f'Q2Position==={Q2Position}')

# Q2 Class

Q2Class = [] # Last index is used to find the one before that

print(f'Taking Rounded Q2 Index')
for Check in range(1,len(data)):
    if Q2Position < data[Check][3]: # Check position with respects to cumulative value
        print(f'Q2 is in {data[Check][0]} Class')
        Q2Class = data[Check]
        Q2Class.append(Check) # Last index is used to find the one before that
        break
        
print(f'Q2Class===={Q2Class}')

'''
LINEAR INTERPOLATION VIA RATIOS

b1===================
-0.5 ----- Q2 ----- 5.5
a1=========

b2====================
0 ----- n/4 ----- 26
a2=========

a1/b1 = a2/b2

the two ratios are the same because the items are assumed to be spread out evenly throughout the whole class

a1 = from Q2 class's lower bound to Q2 --> Q2 - lower bound
b1 = from Q2 class's lower bound to Q2 class's Upper bound --> Upper bound - Lower Bound
a2 = 0th Item to Q2 Item (position wise), Cumulative of lower class to Q2's position --> Q2 Position - cumu of Q2 -1 class
b2 = Cumu of Q2 class -1 to Cumu of Q2 Class --> Cumu Q2 - (Cumu Q2 -1 class)

Q2 - Q2Lb / Q2Ub - Q2Lb = (Q2Position - CumuQ2-1 ) / (CumuQ2 - CumuQ2-1)

Q2 - aOneWithoutQ2 / bOne = aTwo / bTwo

Q2 - aOneWithoutQ2 = ( aTwo / bTwo ) * bOne

Q2 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ2

'''



CumuQ2Minus1 = 0

if Q2Class[4] == 1:
    #This means the Q2 class was the first class
    CumuQ2Minus1 = 0 # Start is 0
else:
    CumuQ2Minus1 = data[Q2Class[4] -1][3]

aOneWithoutQ2 = Q2Class[1][0]
bOne = Q2Class[1][1] - Q2Class[1][0]
aTwo = Q2Position -  CumuQ2Minus1 # Q2Class[4] has the index of the Q2 class, -1 is the one before that
bTwo = Q2Class[3] -  CumuQ2Minus1

# print(aTwo)

Q2 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ2

print(f'Q2 === {Q2}')
```

```python
# Find Q3

import math

Q3Position = math.ceil(3 * (data[-1][3]) / 4) # n

print(f'Q3Position=={Q3Position}')

# Q1 Class

Q3Class = [] # Last index is used to find the one before that

print(f'Taking Rounded Q1 Index')
for Check in range(1,len(data)):
    if Q3Position < data[Check][3]: # Check position with respects to cumulative value
        print(f'Q3 is in {data[Check][0]} Class')
        Q3Class = data[Check]
        Q3Class.append(Check) # Last index is used to find the one before that
        break
        
print(f'Q3Class==={Q3Class}')

'''
LINEAR INTERPOLATION VIA RATIOS

b1===================
10.5 ----- Q3 ----- 15.5
a1=========

b2====================
43 ----- 3n/4 ----- 54
a2=========

a1/b1 = a2/b2

the two ratios are the same because the items are assumed to be spread out evenly throughout the whole class

a1 = from Q3 class's lower bound to Q3 --> Q3 - lower bound
b1 = from Q3 class's lower bound to Q3 class's Upper bound --> Upper bound - Lower Bound
a2 = 43rd Item to Q3 Item (position wise), Cumulative of lower class to Q3's position --> Q3 Position - cumu of Q3 -1 class
b2 = Cumu of Q3 class -1 to Cumu of Q3 Class --> Cumu Q3 - (Cumu Q3 -1 class)

Q3 - Q3Lb / Q3Ub - Q3Lb = (Q3Position - CumuQ3-1 ) / (CumuQ3 - CumuQ3-1)

Q3 - aOneWithoutQ3 / bOne = aTwo / bTwo

Q3 - aOneWithoutQ3 = ( aTwo / bTwo ) * bOne

Q3 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ3

'''



CumuQ3Minus1 = data[Q3Class[4] -1][3]


aOneWithoutQ3 = Q3Class[1][0]
bOne = Q3Class[1][1] - Q3Class[1][0]
aTwo = Q3Position -  CumuQ3Minus1 # Q3Class[4] has the index of the Q3 class, -1 is the one before that
bTwo = Q3Class[3] -  CumuQ3Minus1

# print(aTwo)

Q3 = ( ( aTwo / bTwo ) * bOne ) + aOneWithoutQ3

print(f'Q3 === {Q3}')
```

```python
InterquartileRange = Q3 - Q1

print(InterquartileRange)
```

```python
'''
Anything that falls outside of the fence is an outlier

Lower fence = Q1 - (1.5 * IQR)

Upper fence = Q3 + (1.5 * IQR) 
'''
LowerFence = Q1 - (1.5 * InterquartileRange)
UpperFence = Q3 + (1.5 * InterquartileRange)

print(LowerFence)
print(UpperFence)
```

```python
import matplotlib.pyplot as plt

# Manually calculated values (from your earlier data)
Q1 = Q1
Q3 = Q3
Median = Q2   # you can adjust this based on actual value
LowerFence = LowerFence
UpperFence = UpperFence
Min = 0        # actual minimum (within fences)
Max = data[-1][1][1]     # actual maximum (within fences)

plt.figure(figsize=(8, 2))

# Draw box (from Q1 to Q3)
plt.hlines(1, Q1, Q3, color='skyblue', linewidth=10)

# Draw median
plt.vlines(Median, 0.9, 1.1, color='red', linewidth=2)

# Draw whiskers (fences to actual min/max within range)
plt.hlines(1, Min, Q1, color='black', linestyle='--')
plt.hlines(1, Q3, Max, color='black', linestyle='--')

# Draw end caps (min and max)
plt.plot([Min, Max], [1, 1], 'ko')  # black dots

# Optionally draw fences
plt.axvline(LowerFence, color='gray', linestyle=':', label='Lower Fence')
plt.axvline(UpperFence, color='gray', linestyle=':', label='Upper Fence')

plt.yticks([])
plt.title("Manually Created Box Plot")
plt.xlabel("Height")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
